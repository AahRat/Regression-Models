```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
mpg = sns.load_dataset("mpg")
```

```python
mpg.dropna(inplace=True)
```

```python
sns.scatterplot(data=mpg, x="horsepower", y="mpg")
```

```python
mpg.head()
```

```python
x = mpg.iloc[:,3]
```

```python
x = x.values
```

```python
y = mpg.iloc[:,0]
```

```python
x = x.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
polynomialFeatures = PolynomialFeatures(degree=2)
```

```python
xTrainPoly = polynomialFeatures.fit_transform(x_train)
```

```python
xTestPoly = polynomialFeatures.fit_transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
polyModel = LinearRegression()
```

```python
polyModel.fit(xTrainPoly,y_train)
```

```python
yPredict = polyModel.predict(xTestPoly)
```

```python
yPredict
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(y_test,yPredict)
```

```python
linearModel = LinearRegression()
```

```python
linearModel.fit(x_train,y_train)
```

```python
y_linPredict = linearModel.predict(x_test)
```

```python
mean_absolute_error(y_test,y_linPredict)
```
